```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def fetch_html_tables(url):
    "Returns a list of tables in the html of url"
    page = requests.get(url)
    bs=BeautifulSoup(page.content)
    tables=bs.find_all('table')
    return tables

tables=fetch_html_tables('https://www.motor.no/aktuelt/motors-store-vintertest-av-rekkevidde-pa-elbiler/217132')
table_html=tables[0]

#print(table_html)

#henter inn datasettet som vist i: 9 - webskraping med python.
```

```python
def html_to_table(html):
    "Returns the table defined in html as a list"

    table=[]

    for row in html.find_all('tr'):
        r=[]

        cells=row.find_all('td')
        

        if len(cells)==0:
            cells=row.find_all('th')
            

        for cell in cells:
            cell=format(cell)
            r.append(cell)
        

        table.append(r)
    return table

def format(cell):
    "Returns a string after converting bs4 object cell to clean text"
    if cell.content is None:
        s=cell.text
    elif len(cell.content)==0:
        return ''
    else:
        s=' '.join([str(c) for c in cell.content])
        

    s=s.replace('\xa0','')
    s=s.replace('\n','')
    return s

table=html_to_table(table_html)

#print(str(table)[:1000])
```

```python
';'.join(table[0])
```

```python
def save_data(file_name,table):
    "Saves table to file_name"
    f=open(file_name,'w')
    for row in table:
        f.write(';'.join(row)+'\n')
    f.close()
    
save_data('cars.csv',table)
```

```python
cars = pd.read_csv('cars.csv', delimiter=';',on_bad_lines='skip', encoding='latin1')
```

```python
cars.columns = ['modell', 'wltptall', 'stopp', 'avvik'] #endrer navn på kolonnene
cars[['km', 'kWh']] = cars["wltptall"].str.split("/", expand = True) #splitter km fra KWh
```

```python
cars['km'] = cars['km'].str.replace(r'km', '')
cars['kWh'] = cars['kWh'].str.replace(r'kWh', '')
cars['stopp'] = cars['stopp'].str.replace(r'km', '')
cars['avvik'] = cars['avvik'].str.replace(r'%', '')
#Fjærner benevninger fra tallene
```

```python
cars = cars.replace('x','0', regex=True) #erstatter x med 0
cars[["stopp", "km"]] = cars[["stopp", "km"]].apply(pd.to_numeric) #konverterer de nødvendige tallene til numeric
cars_winter = cars.filter(['modell','stopp','km'], axis=1)#Lagrer de nødvendige kolonnene i et eget datasett

cars_winter
```

```python
fig, ax = plt.subplots()

x = y = np.linspace(0, 700, 100)

ax.scatter(cars_winter['km'], cars_winter['stopp'], label='Observasjoner')

ax.plot(x, y, color='green', label='Forventet prestasjon')

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_xlabel('km')
ax.set_ylabel('stopp')
ax.set_xlim(0, 700)
ax.set_ylim(-30, 700)

plt.title('Rekkeviddetall')

ax.legend();
```
